```python
#import libraries
import pandas as pd
import os

#Get folder path
file_path = '/content/drive/MyDrive/Data/HRDataSet.xlsx'

#data connection
df = pd.read_excel(file_path)
df.head()
```

```python
#Data Analysis
pd.pivot_table(
    df,
    index=['Gender', 'PayType'],
    values=['Salary'],
    aggfunc='sum'
    )
```

```python
pd.pivot_table(
    df,
    index=['Dep', 'Position'],
    values=['Salary'],
    aggfunc='mean'
    )
```

```python
pd.pivot_table(
    df[df['PayType'] == 'Full Time'],
    index=['Age'],
    values=['Years no.'],
    aggfunc='sum'
    )
```
