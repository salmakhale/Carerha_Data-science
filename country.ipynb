```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years=list(map(str,range(1980,2014)))
Algeria=canda.loc['Algeria',years]
Algeria.head()
```

```python
Algeria.plot(kind='line')

plt.title('Algeria')
plt.ylabel('total immigration')
plt.xlabel('years')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(Algeria.index , Algeria.values)

plt.title('Algeria')
plt.ylabel('immigration')
plt.xlabel('years')

plt.xticks(rotation=90)

plt.show()
```

```python
canda_contient = canda.groupby('Continent',axis=0).sum()
canda_contient
```

```python
canda_contient['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')


plt.legend(labels=canda_contient.index,  bbox_to_anchor=(1, 0))
plt.show()
```
