```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd


df= px.data.stocks()

df
```

```python
px.line(df, x='date' , y='MSFT' ,title='Main Title' , labels={'date':'Date' ,'MSFT':'MSFT'})
```

```python
px.line(df, x='date' , y=['GOOG','AAPL'] ,title='Stock Price' , markers =True)
```

```python
fig= go.Figure()

fig.add_trace(
    go.Scatter(x=df['date'],
               y=df['AAPL'],
               mode= 'lines',
               line=dict(color='green', dash ='dash'),
               name='Apple'
                )
    )

fig.add_trace(
    go.Scatter(x=df['date'],
               y=df['GOOG'],
               mode= 'lines+markers',
               line={'color':'yellow' , 'width':2 ,'dash':'dot'},
               name='Google'
                )
    )

fig.add_trace(go.Scatter(x=df['date'],y=df['FB'],line=dict(color='blue', dash ='dot'),name='Facebook'))
```

```python
fig.update_layout(
    title = 'Stocks Variance Over Time',
    xaxis_title = 'date',
    yaxis_title = 'Stocks in $',
    xaxis = {'showgrid' : False},
    yaxis = {'showgrid' : False},
    plot_bgcolor = 'white'
)
```

```python
df = px.data.gapminder()
df = df[df['continent'] == 'Africa']
df2 = pd.pivot_table(df, index = 'country', values = 'pop', aggfunc = 'mean').reset_index()
px.bar(df2, x = 'country', y = 'pop', text = 'pop', color = 'country', text_auto = ',.2s').update_traces(textposition = 'outside')
```

```python
df=px.data.gapminder()

years =df['year'].unique()

app =Dash(__name__)
app.layout=html.Div([
    dcc.Graph(id='gapminder_chart'),
    dcc.RadioItems(
        id='year_selection',
        options=[{'label':str(y),'value':y} for y in years],
        value=years[0]
    )]
)
@app.callback(
    Output(component_id='gapminder_chart',component_property='figure'),
    Input(component_id='year_selection',component_property='value')
)
def update_chart(selected_year):
    filtered_df=df[df['year']==selected_year]
    px.bar(filtered_df,x='country',y='pop',color='continent', text='pop', text_outo=',.2s')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
