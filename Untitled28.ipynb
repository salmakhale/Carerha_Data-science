```python
import pandas as pd
import matplotlib.pyplot as plt
customers = {
    'name': ['Mohamed','Essam','Saad','Mohamed'],
    'age': [20,30,40,20],
    'country': ['Egypt', 'USA', 'Italy','Egypt'],
    'sales': [100,200,600,400],
    'quantity':[200,400,500,20],
    'date':['2024-01-01','2024-01-01','2024-02-01','2024-03-01']
}
# Read Data from Source (Data Connection)
df = pd.DataFrame(customers)
# Data Cleaning & Transformation
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
df['month'] = df['date'].dt.month.astype(str)
```

```python
#column chart
pvt_table = pd.pivot_table(data=df , index='month'  , values=['sales','quantity'] , aggfunc='sum')
```

```python
bw=0.35
x1=pvt_table.index
y1=pvt_table['sales']

x2=[i + bw for i in range(len(pvt_table.index))]
y2=pvt_table['quantity']

plt.bar(x1 , y1, width=bw)
plt.bar(x2 , y2, width=bw)
```

```python
# pie chart
plt.pie(pvt_table['sales'],
        labels = pvt_table.index,
        autopct = '%1.2f%%',
        explode = [0.2, 0, 0]
        );
plt.title('Main Title')
```
