```python
#import libraries
import pandas as pd
import os

#Get folder path
folder_path ='/content/drive/MyDrive/Data/Sales'

#Get files in folder
files = os.listdir(folder_path)

#Get to read each file seperately
dfs =[]
for filename in files :
  #create flie_path
  file_path = folder_path + '/' + filename
  df = pd.read_excel(file_path)
  dfs.append(df)

#combined all Dataframs into one df
combined_df = pd.concat(dfs, ignore_index='True')
combined_df
```

```python
pd.pivot_table(
    combined_df,
    index= 'City' ,
    columns= 'StoreID',
    values='Sales',
    aggfunc='sum'
)
```
