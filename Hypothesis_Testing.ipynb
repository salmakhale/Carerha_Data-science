```python
# Hypothesis Testing (Validate True/False)

import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

np.random.seed(42)

sample_df = pd.DataFrame({

    'id' : range(1,1001),

    'age': np.random.randint(17,80, size = 1000),

    'gender': np.random.choice(['Male', 'Female'], size = 1000),

    'income': np.random.normal(50000, 10000, size = 1000)

})

sample_df
```

```python
hypothesis_paramater = 60000

income_mean = sample_df['income'].mean()

income_std = sample_df['income'].std()

z_score = (hypothesis_paramater - income_mean) / income_std

print(f'z_score = {z_score}')

from scipy.stats import norm

p_value = norm.cdf(z_score, loc = 0, scale = 1)

print(f'p_value = {p_value}')

if p_value < 0.05:

  print('Reject H0')

else:

  print('Failed to Reject H0')
```

```python
male_income = sample_df[sample_df['gender'] == 'Male']['income']

female_income = sample_df[sample_df['gender'] == 'Female']['income']

from scipy.stats import ttest_ind

t_score, p_value = ttest_ind(male_income, female_income)

if p_value < 0.05:

  print('Reject H0')

else:

  print('Failed to Reject H0')
```
